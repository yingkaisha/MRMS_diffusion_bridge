```python
year = 2023
filename = '/glade/campaign/cisl/aiml/ksha/GFS/MRMS_01H_y{}_025.hdf'.format(year)

with h5py.File(filename, 'r') as h5io:
    MRMS025 = h5io['MRMS'][...]
```

```python
N_days = 300
grid_shape = lon_GFS.shape
```

```python
LEADs = np.arange(3, 168+3, 3)
#LEADs = [3,]

save_name = 'MRMS_3H_lead{:02d}_2023.hdf'

MRMS_lead = np.empty((N_days,)+grid_shape); 

for l, lead in enumerate(LEADs):
    MRMS_lead[...] = np.nan
    
    # hourly to 3-hourly
    for d in range(N_days):
        h1 = d*24 + lead
        h0 = h1-3
        MRMS_lead[d, ...] = np.sum(MRMS025[h0:h1, ...], axis=0)
    
    save_name_ = save_name.format(lead)
    print('complete: {}'.format(save_name_))
    
    tuple_save = (MRMS_lead,)
    label_save = ['MRMS_lead',]
    du.save_hdf5(tuple_save, label_save, '/glade/campaign/cisl/aiml/ksha/GFS/', save_name_)
```

```python
LEADs = np.arange(6, 168+6, 6)
save_name = 'MRMS_6H_lead{:02d}_2023.hdf'

MRMS_lead = np.empty((N_days,)+grid_shape); 

for l, lead in enumerate(LEADs):

    MRMS_lead[...] = np.nan

    # hourly to 3-hourly
    for d in range(N_days):
        h1 = d*24 + lead
        h0 = h1-6
        MRMS_lead[d, ...] = np.sum(MRMS025[h0:h1, ...], axis=0)
    
    save_name_ = save_name.format(lead)
    print('complete: {}'.format(save_name_))
    
    tuple_save = (MRMS_lead,)
    label_save = ['MRMS_lead',]
    du.save_hdf5(tuple_save, label_save, '/glade/campaign/cisl/aiml/ksha/GFS/', save_name_)
```
