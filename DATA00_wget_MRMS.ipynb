## Fill-in-the-gap version

```python
# check if files exist
date_list_fill = []

for dt in date_list:
    dt_str = datetime.strftime(dt, '%Y%m%d%H')
    filename = '/glade/campaign/cisl/aiml/ksha/wget_MRMS/{}_done.txt'.format(dt_str)
    
    if os.path.isfile(filename):
        #print('{} Exists'.format(save_name))
        #print(filename)
        continue;
    else:
        date_list_fill.append(dt)
    
```

```python
L_fill = len(date_list_fill)
print('need to download {} files'.format(L_fill))
#date_list_fill = date_list_fill[::-1]
```

```python
L = int(L_fill/10)+1

N = 9
for i in range(N*L, (N+1)*L, 1):
    dt = date_list_fill[i]
    dt_str = datetime.strftime(dt, '%Y%m%d%H')
    dt_folder = datetime.strftime(dt, '%Y/%m/%d/')
    print('wget -L -O {}.zip https://mrms.agron.iastate.edu/{}{}.zip'.format(dt_str, dt_folder, dt_str))
    print('unzip {}.zip'.format(dt_str))
    print('cp {}/CONUS/MultiSensor_QPE_01H_Pass2/MRMS* .'.format(dt_str))
    print('cp {}/CONUS/MultiSensor_QPE_03H_Pass2/MRMS* .'.format(dt_str))

    print('cp {}/CONUS/MountainMapper_QPE_01H/MRMS* .'.format(dt_str))
    print('cp {}/CONUS/MountainMapper_QPE_03H/MRMS* .'.format(dt_str))
    
    print('cp {}/CONUS/RadarAccumulationQualityIndex_01H/MRMS* .'.format(dt_str))
    print('cp {}/CONUS/RadarAccumulationQualityIndex_03H/MRMS* .'.format(dt_str))
    print('rm -rf {}.zip; rm -rf {}'.format(dt_str, dt_str))
    print('touch {}_done.txt'.format(dt_str))
```
